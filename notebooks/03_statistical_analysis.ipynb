## Import Library

```python
import pandas as pd
from scipy import stats
from statsmodels.stats.proportion import proportions_ztest
```

## Load Data

```python
df = pd.read_csv('../data/raw/ab_test_data.csv')
```

## Pisahkan Grup

```python
control = df[df['group'] == 'Control']
treatment = df[df['group'] == 'Treatment']
```

## UJI 1: CONVERSION RATE (Z-Test)

```python
# Hitung jumlah user yang convert (beli)
successes = [control['is_converted'].sum(), treatment['is_converted'].sum()]
# Hitung total user per grup
nobs = [len(control), len(treatment)]

z_stat, p_val_cr = proportions_ztest(successes, nobs)

print(f"--- Conversion Rate ---")
print(f"Control CR  : {control['is_converted'].mean():.4f}")
print(f"Treatment CR: {treatment['is_converted'].mean():.4f}")
print(f"P-Value     : {p_val_cr:.4f}")
```

## UJI 2: REVENUE PER USER (Welch's T-Test)

```python
# equal_var=False membuat ini menjadi Welch's T-Test
t_stat, p_val_rev = stats.ttest_ind(control['revenue'], treatment['revenue'], equal_var=False)

print(f"\n--- Revenue per User ---")
print(f"Control Rev  : {control['revenue'].mean():.2f}")
print(f"Treatment Rev: {treatment['revenue'].mean():.2f}")
print(f"P-Value      : {p_val_rev:.4f}")
```
